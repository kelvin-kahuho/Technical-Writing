```python
#Importing necessary files
import matplotlib.pyplot as plt
```

**Basic Plotting**

To create a simple plot, you'll need some data. For this example, we'll create a basic line plot:


```python
# Sample data
x = [1, 2, 3, 4, 5, 18, 25, 45]
y = [2, 4, 6, 8, 10, 18, 36, 42]

# Create a line plot
plt.plot(x, y)

# Show the plot
plt.show()
```

```python
# Sample data
x = [1, 2, 3, 4, 5, 18, 25, 45]
y = [2, 4, 6, 8, 10, 18, 36, 42]

# Create a line plot with customizations
plt.plot(x, y, marker='x', linestyle='-.', color='b', label='My Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('My First Matplotlib Plot')
plt.legend()

# Save the plot as a PNG file
plt.savefig('my_plot.png')

# Show the plot
plt.show()
```
